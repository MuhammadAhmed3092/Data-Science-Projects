```python
# Select a single league and season
selected_league = 'Eng1'  # Replace with your desired league name
selected_season = 20  # Replace with your desired season

# Filter the dataset for the selected league and season
league_df = df[(df['League'] == selected_league) & (df['Season'] == selected_season)].copy()

# Create a new column to represent the result of each match
league_df['Result'] = league_df['Winner'].apply(lambda x: 'W' if x == 'H' else 'L' if x == 'A' else 'D')

# Create a dictionary to store the points for each team
team_points = {}

# Iterate over each match in the selected league and season
for _, row in league_df.iterrows():
    home_team = row['Home Team']
    away_team = row['Away Team']
    result = row['Result']

    # Update points for the home team
    if home_team not in team_points:
        team_points[home_team] = 0
    if result == 'W':
        team_points[home_team] += 3
    elif result == 'D':
        team_points[home_team] += 1

    # Update points for the away team
    if away_team not in team_points:
        team_points[away_team] = 0
    if result == 'L':
        team_points[away_team] += 3
    elif result == 'D':
        team_points[away_team] += 1

# Create the points table DataFrame
points_table = pd.DataFrame({'Team': list(team_points.keys()), 'Points': list(team_points.values())})

# Sort the points table by points in descending order
points_table = points_table.sort_values('Points', ascending=False).reset_index(drop=True)

print(points_table)
```

```python
#grouping the dataset of Season and Home Goals to find the cummulative sum
#to find home goals for each match
cumulative_home_goals = df.groupby('Season')['Home Goals'].cumsum()
plt.figure(figsize=(20,8))

#this line will create a filled area plot
plt.fill_between(range(len(df)), cumulative_home_goals, alpha=0.3)
plt.xlabel('Match')
plt.ylabel('Cumulative Home Goals')
plt.title('Cumulative Home Goals over Matches')
plt.show()
```

```python
#creating a Frequency relation with that corresponding season
#grouping the Season and Winner columns and calculating frequency in a reshaped grouped data
outcome_counts = df.groupby(['Season', 'Winner']).size().unstack()
plt.figure(figsize=(20,8))

#stacked=true means that bars are stacked on top of each other
outcome_counts.plot(kind='bar',stacked=True)
plt.xlabel('Season')
plt.ylabel('Frequency')
plt.title('Outcome Frequency By Season')
plt.show()
```

```python
from tabulate import tabulate
# Select a single league and season
selected_league = 'Eng1'  # Replace with your desired league name
selected_season = 23  # Replace with your desired season

# Filter the dataset for the selected league and season
league_df = df[(df['League'] == selected_league) & (df['Season'] == selected_season)].copy()

# Create a new column to represent the result of each match
league_df['Result'] = league_df['Winner'].apply(lambda x: 'W' if x == 'H' else 'L' if x == 'A' else 'D')

# Create a dictionary to store the points for each team
team_points = {}

# Iterate over each match in the selected league and season
for _, row in league_df.iterrows():
    home_team = row['Home Team']
    away_team = row['Away Team']
    result = row['Result']

    # Update points for the home team
    if home_team not in team_points:
        team_points[home_team] = {'Wins': 0, 'Draws': 0, 'Losses': 0,'Points': 0}
    if result == 'W':
        team_points[home_team]['Points'] += 3
        team_points[home_team]['Wins'] += 1
    elif result == 'D':
        team_points[home_team]['Points'] += 1
        team_points[home_team]['Draws'] += 1
    else:
        team_points[home_team]['Losses'] += 1

    # Update points for the away team
    if away_team not in team_points:
        team_points[away_team] = {'Wins': 0, 'Draws': 0, 'Losses': 0,'Points': 0}
    if result == 'L':
        team_points[away_team]['Points'] += 3
        team_points[away_team]['Wins'] += 1
    elif result == 'D':
        team_points[away_team]['Points'] += 1
        team_points[away_team]['Draws'] += 1
    else:
        team_points[away_team]['Losses'] += 1

# Create the points table DataFrame
points_table = pd.DataFrame(team_points).transpose().reset_index().rename(columns={'index': 'Team'})

# Sort the points table by points and other criteria
points_table = points_table.sort_values(['Points', 'Wins', 'Team'], ascending=[False, False, True]).reset_index(drop=True)

# Set the display options for pandas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Create the table title
table_title = 'English Premier League 22/23'
# Display the points table using tabulate
print(table_title)
print(tabulate(points_table, headers='keys', tablefmt='github', showindex=True))
```
