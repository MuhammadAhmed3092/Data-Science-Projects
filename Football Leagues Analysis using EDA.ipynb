```python
bold = "\033[1m"
reset = "\033[0m"
for col in df.columns:
    print(f"{bold}Value Counts of {col}")
    print(f"{bold}Column : {col}{reset}")
    print(df[col].value_counts())
    print("\n")
```

```python
df.head(10)
```

```python
df['Winner Team'] = ''
```

```python
df.loc[df['Winner'] == 'H', 'Winner Team'] = df['Home Team']
df.loc[df['Winner'] == 'A', 'Winner Team'] = df['Away Team']
df.loc[df['Winner'] == 'D', 'Winner Team'] = 'Draw'
```

```python
df['Half-Time Winner'] = ''
df.loc[df['HT Winner'] == 'H', 'Half-Time Winner'] = df['Home Team']
df.loc[df['HT Winner'] == 'A', 'Half-Time Winner'] = df['Away Team']
df.loc[df['HT Winner'] == 'D', 'Half-Time Winner'] = 'Draw'
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
df['Winner'].value_counts().plot.pie(autopct = '%1.2f%%')
plt.title('Winner of match at Full Time')
plt.subplot(1,2,2)
df['HT Winner'].value_counts().plot.pie(autopct = '%1.2f%%')
plt.title("Winner of match at Half Time")
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
df['Total Goals'].value_counts().plot.pie(autopct='%1.2f%%')
plt.title('Total Goals at Full Time')
plt.subplot(1,2,2)
df['FH Goals'].value_counts().plot.pie(autopct='%1.2f%%')
plt.title('Total Goals at Half-Time')
```

```python
winner_counts = df['Winner Team'].value_counts().head(11)
winner = winner_counts[1:]
plt.figure(figsize=(20,8))
winner.plot(kind='bar')
```
