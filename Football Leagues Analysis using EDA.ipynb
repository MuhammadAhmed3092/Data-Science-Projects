```python
df['Winner Team'] = ''
```

```python
df.loc[df['Winner'] == 'H', 'Winner Team'] = df['Home Team']
df.loc[df['Winner'] == 'A', 'Winner Team'] = df['Away Team']
df.loc[df['Winner'] == 'D', 'Winner Team'] = 'Draw'
```

```python
df['Half-Time Winner'] = ''
df.loc[df['HT Winner'] == 'H', 'Half-Time Winner'] = df['Home Team']
df.loc[df['HT Winner'] == 'A', 'Half-Time Winner'] = df['Away Team']
df.loc[df['HT Winner'] == 'D', 'Half-Time Winner'] = 'Draw'
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
df['Winner'].value_counts().plot.pie(autopct = '%1.2f%%')
plt.title('Winner of match at Full Time')
plt.subplot(1,2,2)
df['HT Winner'].value_counts().plot.pie(autopct = '%1.2f%%')
plt.title("Winner of match at Half Time")
```

```python
winner_counts = df['Winner Team'].value_counts().head(11)
winner = winner_counts[1:]
plt.figure(figsize=(20,8))
winner.plot(kind='bar')
plt.title('Top 10 Winner Teams',fontsize=25)
plt.xlabel('Winner Teams',fontsize=20)
plt.ylabel('Number Of Wins',fontsize=20)
plt.xticks(rotation=90)
plt.legend(fontsize=15)
plt.show()
```

```python
htwinner_counts = df['Half-Time Winner'].value_counts().head(11)
htwinner = winner_counts[1:]
plt.figure(figsize=(20,8))
htwinner.plot(kind='bar')
plt.title('Top 10 Half Time Winner Teams',fontsize=25)
plt.xlabel('Half-Time Winner Teams',fontsize=20)
plt.ylabel('Number Of Wins',fontsize=20)
plt.xticks(rotation=90)
plt.legend(fontsize=15)
plt.show()
```

```python
outcomes = ['Home Wins','Draw','Away Wins']
counts = [df['Winner'].value_counts()['H'], df['Winner'].value_counts()['D'], df['Winner'].value_counts()['A']]
plt.title("Distribution of Match Outcomes")
plt.bar(outcomes,counts)
plt.xlabel("Result")
plt.ylabel("Number of Wins")
plt.show()
```

```python
cleansheets_teams=df[['Home Team','Away Team','Home Goals','Away Goals']]
clean_sheets = df[(df['Away Goals'] == 0) & (df['Home Goals'] == 0)]
clean_sheets_by_team = clean_sheets.groupby('Away Team')['Away Team'].count().add(clean_sheets.groupby('Home Team')['Home Team'].count(), fill_value=0)
clean_sheets_by_team = clean_sheets_by_team.sort_values(ascending=False)

top_10_teams = clean_sheets_by_team.head(10)

plt.figure(figsize=(20,8))
plt.bar(top_10_teams.index, top_10_teams)
plt.xlabel('Teams',fontsize=20)
plt.ylabel('Number of Clean Sheets',fontsize=20)
plt.title('Top 10 Teams with the Most Clean Sheets',fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
outcomes = ['Home Win','Draw','Away Win']
counts = [df['Winner'].value_counts()['H'], df['Winner'].value_counts()['D'], df['Winner'].value_counts()['A']]
plt.title("Distribution of Match Outcomes")
plt.pie(counts, labels=outcomes, autopct='%1.1f%%')
plt.show()
```

```python
seasons = df['Season']
home_goals = df['Home Goals']
away_goals = df['Away Goals']

plt.plot(seasons, home_goals, label = 'Home Goals')
plt.plot(seasons, away_goals, label = 'Away Goals')
plt.xlabel('Season')
plt.ylabel('Goals')
plt.title('Goals Scored by Home and Away Teams over the season')
plt.legend()
plt.show()
```

```python
total_goals = df['Total Goals']

plt.hist(total_goals, bins=10)
plt.xlabel('Total Goals')
plt.ylabel('Times x amount of goals scored by a team')
plt.title('Distribution of Total Goals')
plt.show()
```

```python
home_goals = df['Home Goals']
away_goals = df['Away Goals']

plt.scatter(home_goals, away_goals)
plt.xlabel('Home Goals')
plt.ylabel('Away Goals')
plt.title('Relationship b/w Home and Away goals')
plt.show()
```

```python
teams = df['Home Team'].unique()
average_goals = []
for team in teams:
    home_avg = np.mean(df[df['Home Team'] == team]['Home Goals'])
    away_avg = np.mean(df[df['Away Team'] == team]['Away Goals'])
    average_goals.append((home_avg + away_avg) / 2)

plt.figure(figsize=(20,8))    
plt.bar(teams, average_goals)
plt.xlabel('Team')
plt.ylabel('Average Goals')
plt.title('Average Goals Scored by Each Team')
plt.show()
```
